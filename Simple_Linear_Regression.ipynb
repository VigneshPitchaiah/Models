```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total_spend'] = df['TV'] + df ['radio'] + df ['newspaper']
```

```python
df.head()
df.tail()
```

```python
sns.scatterplot(data=df, x='Total_spend',y='sales')
```

```python
X = df['Total_spend']
y = df['sales']
```

```python
# y = mx + b
# y = B1x + Bo

arr = np.polyfit(X,y,deg = 1)
print(arr)
print(arr[1])
```

```python
predicted_sales = arr[0]*X + arr[1]
```

```python
predicted_sales
```

```python
sns.scatterplot(x='Total_spend',y='sales', data=df)
plt.plot(X,predicted_sales,color='r')
```

```python
sns.regplot(data=df, x='Total_spend',y='sales')
```

```python
spend = 200

predicted_sales = arr[0]*spend + arr[1]
print(predicted_sales)
```

```python
line = np.polyfit(X,y,3)
print(line)
```

```python
#Polynomial 3rd degree
X = np.linspace(0,700,200)
predict_sales = line[0]*X**3 + line[1]*X**2 + line[2]*X + line[3]
```

```python
sns.scatterplot(x='Total_spend',y='sales', data=df)
plt.plot(X,predict_sales,color='r')
```
